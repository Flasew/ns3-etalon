```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
font = {'family' : 'sans',
        'weight' : 'normal',
        'size'   : 20}
matplotlib.rc('font', **font)
```

```python
def plot_lat_cdf(ax, df, dt, ql, dl, nf, path): 
    sns.set_palette(sns.color_palette("husl", len(nf)))
    ddf = df[(df['ql']==ql)&(df['delay']==dl)&(df['dt']==dt)]
    for i in nf:
        sdf = ddf[ddf['nflows']==i]   
        fname =  path + "/httplat_" + sdf['tstamp'].iloc[0] + ".csv"
        ldata = pd.read_csv(fname, header=None, index_col=None, names=['time', 'latency'])
        ldata['latency'] =ldata['latency'].apply(lambda x: int(x/1000)) 
        ax = ldata['latency'].plot.hist(ax=ax, density=True, cumulative=True, label='CDF',
             histtype='step', alpha=0.8, bins=100000, figsize=(18,12), linewidth=3)
    return ax
        
        
```

```python
pdp = "log_httpranddelayalllog/pdata"
pdata = pd.read_csv(pdp, sep='\t', header=None, index_col=None,
                    names=['tstamp', 'dt', 'nflows', 'ql', 'delay', 'fct', 'retrans', 'rto', 'syn', 'rtt'])
pdata['delay'] = pdata['delay'].apply(lambda x: int(x/2000))
# pavg = (pdata
#         .groupby(['nflows', 'dt', 'ql', 'delay'], as_index=False)
#         .agg({'fct': np.mean, 'retrans': np.mean, 'rto': np.mean, 'rtt': np.mean, "syn": np.mean})
#        )
```

```python
plot_lat_cdf(None, pdata, 65536, 400, 2000, [100, 200, 400, 800], "log_httpranddelayalllog")
```
